## Extract information from the given web site

You will extract the data from the below web site: <br>


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
from bs4 import BeautifulSoup 
import requests
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
datacsv=[]
```

Download the webpage at the url


```python
#your code goes here
data  = requests.get(url).text 
```

Create a soup object


```python
#your code goes here
soup = BeautifulSoup(data,"html5lib")
```

Scrape the `Language name` and `annual average salary`.


```python
data  = requests.get(url).text 
soup = BeautifulSoup(data,"html5lib")
table = soup.find('table')
for row in table.find_all('tr'): 

    cols = row.find_all('td') 

    programing_language = cols[1].getText()

    salary = cols[3].getText() 
    datacsv.append([programing_language,salary])

    

    print("{}--->{}".format(programing_language,salary))

```

Save the scrapped data into a file named *popular-languages.csv*


```python
cols=['programing_language','salary']
df = pd.DataFrame(datacsv,columns=cols)
df.to_csv("popular-languages.csv", index=False)
print(df)
df_sorted_desc= df.sort_values('salary',ascending=False)
plt.figure(figsize=(10,6))
# bar plot with matplotlib
plt.bar('programing_language', 'salary',data=df_sorted_desc)
plt.xlabel("programing_language", size=15)
plt.ylabel("Salary in US Dollars", size=15)
plt.title("Bar plot in Descending Order with Matplotlib", size=18)
plt.savefig("bar_plot_matplotlib_descending_order_Python.png")

```
